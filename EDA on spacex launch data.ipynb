**2. Checking for Missing Values and Duplicated Values and fill them.**

```python
 # filling the Flight Number.....
for i in df.index:
    df.at[i,'Flight Number'] = str(i+1)
```

```python
df.isnull().sum()
```

```python
# fill the date coloum.....
df['Date'].fillna(df['Date'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill the Booster Version coloum.....
df['Booster Version'].fillna(df['Booster Version'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# Transform the data to int because in coloumn weight are in inteager 
# but the datatype is string thats why i convert it to int
digit_list = []
for i in np.array(list(df['Payload Mass (kg)'])) :
    digit = ''
    for k in i:
        if k.isdigit():
            digit += k
    if not len(digit) == 0:   
        digit_list.append(int(digit))
        
# fill the Payload Mass coloum.....
df['Payload Mass (kg)'].fillna(round(np.mean(digit_list)),inplace=True)
```

```python
# There are two value name as classified.......convert this to a mass...
for i,j in enumerate(df['Payload Mass (kg)']):
    if not str(j).isdigit():
        df.at[i,'Payload Mass (kg)'] = round(np.mean(digit_list))
```

```python
df.isnull().sum()
```

```python
# Fill Customer null value....
df['Customer'].fillna(df['Customer'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill the null value of mission outcome.
df['Mission Outcome'].fillna(df['Mission Outcome'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill the null value of Landing outcome.
df['Landing Outcome'].fillna(df['Landing Outcome'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

**The Countplots Shows The Orbit, Launch Site, Mission Outcome And Landing Outcome Distributions In The Dataset**

```python
df.head(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 16))

sns.countplot(x='Orbit', data=df, ax=axes[0, 0],hue='Orbit')
axes[0, 0].set_title('Orbital Distribution')
axes[0, 0].tick_params(axis='x', rotation=90)

sns.countplot(x='Launch Site', data=df, ax=axes[0, 1],hue='Launch Site')
axes[0, 1].set_title('Launch Site Distribution')
axes[0, 1].tick_params(axis='x', rotation=90)

sns.countplot(x='Mission Outcome', data=df, ax=axes[1, 0],hue='Mission Outcome')
axes[1, 0].set_title('Mission Outcome Distribution')
axes[1, 0].tick_params(axis='x', rotation=90)

sns.countplot(x='Landing Outcome', data=df, ax=axes[1, 1],hue='Landing Outcome')
axes[1, 1].set_title('Landing Outcome Distribution')
axes[1, 1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

**Scatter Plot Between Time With Landing Outcome**
